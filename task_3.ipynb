### Visualizing Data

```python
sns.countplot(x='Ship Mode', data=df)
plt.show()
```

```python
sns.countplot(x='Segment', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.title('State wise shipments')
plt.show()
```

### As we can see, California has highest sales.Texas, New York, etc are active in business.Few states like North Dakota, Wyoming and West Virginia have less or no sales at all.
### To generate more profit we have to increase our sales in Texas and New York where we have potential customers.

```python
plt.figure(figsize = (5,5))
textprops = {"fontsize":12}
plt.title('Category')
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

#### Office supplies have highest demand

```python
df.groupby('Region')['Profit'].agg(sum).plot(kind='bar')
plt.ylabel('Profit')
plt.title('Profits in different regions')
plt.show()
```

#### We can see that west generates most profit. Central and Southern region are generating less profit as compared to the Eastern and Western Region.We need to increase the sales in Central and Southern region in order to generate more profit in these region.

```python
rt = df.groupby('State')['Profit'].sum()
plt.figure(figsize=(15,5))
sns.barplot(x=rt.index, y=rt.values)
plt.xticks(rotation=90)
plt.ylabel('Profit')
plt.show()
```

```python
sns.pairplot(df, hue='Segment')
```

#### Heatmap showing correlation between features

```python
sns.heatmap(df.corr(), annot = True, cmap = "plasma")
```

```python
df.groupby(['Category'])[['Profit','Sales']].agg(sum).plot(kind='bar',figsize=(8,4), )
plt.ylabel('Profit and Sales')
plt.show()
```

#### This shows that Furniture is getting very less profit

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Category',hue='Segment',data=df)
plt.ylabel('Count')
plt.title('Category')
```

#### Consumer segment has highest purchases whereas Home Office segment has the least purchases in each category.Focus should be on the corporate and home office segment to increase our sales and profit.

```python
df.groupby('Sub-Category')[['Sales','Profit']].agg(sum).plot(kind='barh',figsize=(10, 5))
plt.ylabel('Sub-Category')
plt.xlabel('Sales and Profit')
plt.show()
```

#### Sales of Bookcases, Supplies, Tables are facing loss. On the sales of tables we are facing highest loss.

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Region',hue='Ship Mode',data=df)
plt.ylabel('Count of Ship Mode')
plt.title('Count of ship mode by region and ship mode')
```

```python
df.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "plasma")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 14)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.lineplot(df['Discount'], df['Profit'], data=df)
```
